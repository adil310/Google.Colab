```python

import numpy as np
import matplotlib.pyplot as plt

# Step 2: Generate random numbers
random_numbers = np.random.randn(10000)

# Step 3: Create Histogram
plt.figure(figsize=(8, 6))
plt.hist(random_numbers, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Step 4: Create Pie Chart
unique, counts = np.unique(random_numbers, return_counts=True)
plt.figure(figsize=(8, 6))
plt.pie(counts, labels=unique, autopct='%1.1f%%')
plt.title('Pie Chart of Random Numbers')
plt.show()

# Step 5: Create Heatmap (2D Histogram)
plt.figure(figsize=(8, 6))
plt.hist2d(random_numbers, random_numbers, bins=30, cmap='viridis')
plt.colorbar(label='Counts in bin')
plt.title('Heatmap of Random Numbers')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Step 6: Create Line Chart
plt.figure(figsize=(8, 6))
plt.plot(random_numbers)
plt.title('Line Chart of Random Numbers')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
